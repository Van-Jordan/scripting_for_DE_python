```python
from faker import Faker
fake = Faker()
names = [fake.name().split() for i in range(10000)]
names = [name for name in names if len(name) == 2]
names
```

```python
import sqlite3
conn = sqlite3.connect('sample.db')
```

```python
query = 'INSERT INTO users(name,surname) VALUES(?,?)'
cursor = conn.cursor()
for name in names:
    cursor.execute(query, name)
conn.commit()
```

```python
select_query = 'SELECT * FROM users'
for i in cursor.execute(select_query):
    print(i)
```

```python
query = "SELECT distinct name FROM users"
cursor.execute(query)
count = 0
for i in cursor.fetchall():
    print(count, i)
    count = count + 1
```

```python
query = "SELECT name, surname FROM users WHERE name LIKE '%es%'"
for result in cursor.execute(query):
    print(f"First name: {result[0]}, Last name: {result[1]}")
```
